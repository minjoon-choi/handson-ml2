## 훈련 데이터, 테스트 데이터 분리 

### 함수 만들어서 분리 

````python
import numpy as np 

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
````

````python
train_set, test_set = split_train_test(housing, 0.2)
print(train_set.size, test_set.size)
````

````python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
````

````python
housing_with_id = housing.reset_index()

print(housing_with_id.columns)

train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
print(train_set.size, test_set.size)
````

````python
housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"]
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "id")
print(train_set.size, test_set.size)
````

### 계층적 샘플링 stratified sampling

데이터셋이 충분히 크다면 일반적으로 괜찮지만, 그렇지 않다면 샘플링 편향이 생길 가능성! 
- 잘 구성된 설문조사는 샘플에서도 전체 특성 값에서의 비율을 유지해야함 

중간 소득이 중간 주택가격을 예측하는데 매우 중요하다 
- 이 경우 테스트 세트가 전체 데이터셋에 있는 여러 소득 카테고리를 잘 대표해야 함 
- 중간 소득이 연속적인 숫자형 특성이므로 소득에 대한 카테고리 특성을 만들어야 함 

````python
# pd.cut() 함수를 사용해 카테고리 5개를 가진 소득 카테고리 특성을 만듦

housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                              labels=[1, 2, 3, 4, 5])
````

````python
housing["income_cat"].hist()
````

````python
# 소득 카테고리를 기반으로 계층 샘플링 - StratifiedShuffleSplit 사용 

from sklearn.model_selection import StratifiedShuffleSplit 

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
````

````python
# test_set의 소득 카테고리 비율  
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
````

````python
# 전체 dataset의 소득 카테고리 비율 확인 
housing["income_cat"].value_counts() / len(housing)
````

````python
# income_cat 특성을 삭제, 데이터를 원래 상태로 되돌려 둠 

for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True) # axis 0은 행, 1은 열 
````

## 나만의 변환기

특별한 정제 작업이나 특성 조합 등의 작업을 위해 자신만의 변환기를 만들어야 할 때 있음 
- 내가 만든 변환기를 사이킷런의 기능과 매끄럽게 연동하고 싶을 것임 
- 사이킷런은 (상속이 아닌) 덕 타이핑<sup>duck typing</sup>을 지원하므로 fit(), transform(), fit_transform() 메서드를   
  구현한 파이썬 클래스를 만들면 됨  
- 마지막 메서드는 TransformerMixin을 상속하면 자동으로 생성됨 
- 또한 BaseEstimator를 상속하면 하이퍼파라미터 튜닝에 필요한 get_params와 set_params() 메서드를 추가로 얻게 됨 

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BestEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room     
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, 
                         bedroom_per_room]
        
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)

```

## 특성 스케일링 

머신러닝 알고리즘은 입력 숫자 특성들의 스케일이 많이 다르면 잘 작동하지 않음  
- 방 개수의 범위는 6 ~ 39,320 / 중간 소득의 범위는 0~15 
- 타깃값에 대한 스케일링은 일반적으로 불필요 

모든 특성의 범위를 같도록 만들어주는 방법으로 min-max 스케일링과 표준화가 널리 사용됨 

**min-max 스케일링**
- 정규화 normalization 이라고 부름 
- 0 ~ 1 범위에 들도록 값을 이동하고 스케일을 조정
- 데이터에서 최솟값을 뺀 후 최댓값과 최솟값의 차이로 나누면 됨 
- 사이킷런에는 MinMaxScaler 변환기를 제공. feature_range 매개변수로 범위를 0~1에서 변경할 수 있음 

**표준화 standardizaiton**
- 평균을 뺀 후 표준편차로 나누어 결과 분포의 분산이 1이 되도록 함 
- min-max 스케일링과는 달리 범위의 상한과 하한이 없어 어떤 알고리즘에서는 문제가 될 수 있음  
(ex. 신경망은 종종 입력값의 범위로 0~1 사이의 값을 기대) 
- 그러나 표준화는 이상치에 영향을 덜 받는다. 
- 사이킷런에는 StandardScaler 변환기가 있음 

## 변환 파이프라인

예시) 숫자 특성을 처리하는 간단한 파이프라인
```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('attribus_adder', CombinedAttributeAdder()),
        ('std_scaler', StandardScaler()),
    ])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

- Pipeline은 연속된 단계를 나타내는 이름/추정기 쌍의 목록을 입력으로 받음 
- 마지막 단계에는 변환기와 추정기를 모두 사용할 수 있고, 그 외에는 모두 변환기여야 함 
(즉, fit_transform() 메서드를 가지고 있어야 한다.) 
- 이름은 고유하고 이중 밑줄 문자를 포함하지 않는다면 무엇이든 상관 없음
